```python
import pandas
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pandas.read_csv('https://gist.githubusercontent.com/rioto9858/ff72b72b3bf5754d29dd1ebf898fc893/raw/1164a139a780b0826faef36c865da65f2d3573e0/top50MusicFrom2010-2019.csv')
data.head()
```

```python
features = data[['year', 'Beats.Per.Minute -The tempo of the song', 'Energy- The energy of a song - the higher the value, the more energtic', 'Danceability - The higher the value, the easier it is to dance to this song', 'Loudness/dB - The higher the value, the louder the song', 'Liveness - The higher the value, the more likely the song is a live recording', 'Valence - The higher the value, the more positive mood for the song', 'Length - The duration of the song']]
```

```python
item_similarities = cosine_similarity(features)
def get_recommendations(item_id, num_recommendations):
    idx = data.index[data['title'] == item_id].tolist()[0]
    similarity_scores = pandas.Series(item_similarities[idx], index=data['title'])
    sorted_scores = similarity_scores.sort_values(ascending=False)
    recommendations = sorted_scores.index[~sorted_scores.index.isin([item_id])]
    return recommendations[:num_recommendations]
```

```python
item_id = 'TiK ToK'
num_recommendations = 5
recommendations = get_recommendations(item_id, num_recommendations)
print(recommendations)
```
